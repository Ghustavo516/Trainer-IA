```python
categories = list(base_phrases.keys())

def generate_text(category):
    frases = base_phrases[category]
    frase1 = random.choice(frases)
    frase2 = random.choice(frases)
    while frase2 == frase1:
        frase2 = random.choice(frases)
    return f"{frase1} {frase2}"

generated_texts = set()
rows = []

while len(rows) < num_records:
    category = random.choice(categories)
    text = generate_text(category)
    if text not in generated_texts:
        generated_texts.add(text)
        rows.append((text, category))

file_path = "text_classification_finance.csv"
with open(file_path, mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file, quotechar='"', quoting=csv.QUOTE_ALL)
    writer.writerow(["texto", "categoria"])
    writer.writerows(rows)

print(f"Arquivo salvo em {file_path}")
```
